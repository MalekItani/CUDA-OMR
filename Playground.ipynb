```python
import cv2
import numpy as np
```

```python
def code_check():
    # Read image 
    img = cv2.imread('src/one_note.png', cv2.IMREAD_COLOR) # road.png is the filename
    img_shape = 200
    img = cv2.resize(img, dsize=(img_shape, int(img_shape/img.shape[1] * img.shape[0])))
    result = img.copy()
    # Convert the image to gray-scale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, gray = cv2.threshold(gray, 220, 255, cv2.THRESH_BINARY_INV)
    # Find the edges in the image using canny detector
#     edges = cv2.Canny(gray, 50, 200)
#     cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
    # Detect points that form a line
    cv2.imshow("", gray)
    lines = cv2.HoughLinesP(gray, 1, np.pi/2, 100, maxLineGap=15)
    print(lines)
    # Draw lines on the image
    if lines is not None:
        for line in lines:
            x1, y1, x2, y2 = line[0]
            cv2.line(result, (x1, y1), (x2, y2), (255, 0, 0), 1)
    # Show result
    cv2.imshow("Original", img)
    cv2.imshow("Result Image", result)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


def test():
    SRC_PATH = 'src/'
    image_width = 600
    image_height = 800

    filename = SRC_PATH + "one_note.png"
    image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
    # padding_x = (image_width - image.shape[1])//2
    # padding_y = (image_height - image.shape[0])//2
    # image = cv2.copyMakeBorder(image, padding_y, padding_y, padding_x, padding_x, cv2.BORDER_CONSTANT, value=255)
    result = cv2.Sobel(image, cv2.CV_16S, 0, 1, ksize=1)
    result = np.absolute(result).astype(np.uint8)
    
    cv2.imshow("Original", image)
    cv2.imshow("Result", result)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
